```python
#Plotting in Pandas
cutest = pd.Series(np.random.normal(size=15))
print(cutest)
cutest.plot()
```

```python
#Adding grids
cutest.cumsum().plot(grid=True)
```

```python
#using dataframes
pandas = pd.DataFrame({'white': np.random.normal(size=5), 
                       'black': np.random.normal(1,size=5), 
                       'cute': np.random.normal(size=5)})
print(pandas)

pandas.cumsum(0).plot()
```

```python
#Subplots
pandas.cumsum(0).plot(subplots=True)
```

```python
#secondary y-axis
pandas.cumsum(0).plot(secondary_y='white')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(6, 10))
for i,pet in enumerate(['white','black','cute']):
    pandas[pet].cumsum(0).plot(ax=axes[i], title=pet)
axes[0].set_ylabel('y-axis')
```

```python
#saving the figures
fig.savefig('fig.png')
```

```python
#Simple Line Plots
car = plt.figure()
bike = plt.axes()

arun = np.linspace(0, 10, 25)
print(arun)
bike.plot(arun, np.sin(arun));
```

```python
#COLORS
plt.plot(arun, np.sin(arun - 0), color='orange')        
plt.plot(arun, np.sin(arun - 1), color='pink')           
plt.plot(arun, np.sin(arun - 2), color='0.58')       
```

```python
#style
plt.plot(arun, np.sin(arun - 0), linestyle='dashed')        
plt.plot(arun, np.sin(arun - 1), linestyle='dotted')           
plt.plot(arun, np.sin(arun - 2), linestyle='solid')  
```

```python
#Axes Limits
plt.plot(arun, np.sin(arun - 0))

plt.xlim(1, 9)
plt.ylim(-3, 7);
```

```python
#LABELING
plt.plot(arun, np.sin(arun))
plt.title("Frnd")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

```python
aaa = np.random.RandomState(9)
i = aaa.randn(70)
j = aaa.randn(70)
colors = aaa.rand(70)
sizes = 1000 * aaa.rand(70)

plt.scatter(i, j, c=colors, s=sizes, alpha=0.3,
            cmap='rainbow')
plt.colorbar() 
```

```python
#Error bars
p = np.linspace(1, 9, 20)
ab = 0.2
q = np.sin(p) + ab * np.random.randn(20)

plt.errorbar(p, q, yerr=ab, fmt='.b');
```

```python
r = np.random.normal(0, 0.6, 60)
u = np.random.normal(-2, 3, 60)

park = dict(histtype='stepfilled', alpha=0.5, density=True, bins=30)

plt.hist(r, **park)
plt.hist(u, **park)
```

```python
#2-d histograms
s = [6, 3]
t = [[7, 2], [4, 2.5]]
s, t = np.random.multivariate_normal(s, t, 2)
```

```python
plt.hist2d(s, t, bins=20, cmap='gnuplot')
ba = plt.colorbar()
```

```python
plt.hexbin(s, t, gridsize=20, cmap='winter')
ba= plt.colorbar()
```
